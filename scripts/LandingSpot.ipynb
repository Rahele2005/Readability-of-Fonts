```python
def get_word_dimensions(word, font_size):
    # Assuming a monospace font where each character width is the same
    char_width = font_size  # Adjust based on actual font characteristics
    
    # Calculate width and height
    width = len(word) * char_width
    height = font_size  # Assuming font height is same for all characters
    
    return width, height

def measure_sentence(sentence, font_size):
    words = sentence.split()
    word_dimensions = {}
    
    for word in words:
        width, height = get_word_dimensions(word, font_size)
        word_dimensions[word] = (width, height)
    
    return word_dimensions

# Example usage:
font_size = 32  # Example font size (adjust as needed)

sentence = "HI, SARAH! HOW WAS YOUR WEEKEND? WHAT DID YOU DO?"
word_dimensions = measure_sentence(sentence, font_size)

# Print word dimensions
for word, (width, height) in word_dimensions.items():
    print(f"Word '{word}': Width={width}px, Height={height}px")

```

```python
def calculate_sentence_bounding_boxes(sentence, x_start, y_start, word_widths, word_height):
    # Initialize an empty dictionary to store bounding boxes
    bounding_boxes = {}
    
    # Split the sentence into individual words
    words = sentence.split()
    
    # Initialize x-coordinate with starting position
    x = x_start
    
    # Iterate through each word in the sentence
    for word, width in zip(words, word_widths):
        # Calculate bounding box for each word
        x1 = x
        y1 = y_start
        x2 = x + width
        y2 = y_start + word_height
        
        # Store bounding box coordinates for the word
        bounding_boxes[word] = (x1, y1, x2, y2)
        
        # Update x-coordinate for the next word (considering spacing between words)
        x += width  # Add width of the current word
    
    return bounding_boxes

# Example usage:
sentence = "HI, SARAH! HOW WAS YOUR WEEKEND? WHAT DID YOU DO?"
x_start = -288
y_start = 0
word_widths = [96, 192, 96, 128, 128, 224, 128, 96, 96, 96]  # Hypothetical widths for each word
word_height = 32  # Hypothetical height for all words

# Calculate bounding boxes for the sentence
bounding_boxes = calculate_sentence_bounding_boxes(sentence, x_start, y_start, word_widths, word_height)

# Print bounding boxes for each word in the sentence
for word, (x1, y1, x2, y2) in bounding_boxes.items():
    print(f"Bounding box for '{word}': ({x1}, {y1}, {x2}, {y2})")
```

```python
def calculate_sentence_bounding_boxes(sentence, x_start, y_start, word_widths, word_height):
    # Initialize an empty dictionary to store bounding boxes
    bounding_boxes = {}
    
    # Split the sentence into individual words
    words = sentence.split()
    
    # Initialize x-coordinate with starting position
    x = x_start
    
    # Iterate through each word in the sentence
    for word, width in zip(words, word_widths):
        # Calculate bounding box for each word
        x1 = x
        y1 = y_start
        x2 = x + width
        y2 = y_start + word_height
        
        # Store bounding box coordinates for the word
        bounding_boxes[word] = (x1, y1, x2, y2)
        
        # Update x-coordinate for the next word (considering spacing between words)
        x += width  # Add width of the current word
    
    return bounding_boxes

# Example usage:
sentence = "hi, lisa! did you finish the book that i lent you?"
x_start = -288
y_start = 0
word_widths = [96, 160, 96, 96, 192, 96, 128, 32, 128, 128]  # Hypothetical widths for each word
word_height = 32  # Hypothetical height for all words

# Calculate bounding boxes for the sentence
bounding_boxes = calculate_sentence_bounding_boxes(sentence, x_start, y_start, word_widths, word_height)

# Print bounding boxes for each word in the sentence
for word, (x1, y1, x2, y2) in bounding_boxes.items():
    print(f"Bounding box for '{word}': ({x1}, {y1}, {x2}, {y2})")
```

```python
import matplotlib.pyplot as plt

# Fixation coordinates after SENTENCE_UPPERCASE_ONSET
fixation_coordinates_uppercase = {
    'FPOGX': 0.08463,
    'FPOGY': 0.52556
}

# Fixation coordinates after SENTENCE_LOWERCASE_ONSET
fixation_coordinates_lowercase = {
    'FPOGX': 0.09124,
    'FPOGY': 0.55418
}

# Bounding box coordinates for each word in the uppercase sentence
bounding_boxes_uppercase = {
    'HI,': (-288, 0, -192, 32),
    'SARAH!': (-192, 0, 0, 32),
    'HOW': (0, 0, 96, 32),
    'WAS': (96, 0, 224, 32),
    'YOUR': (224, 0, 352, 32),
    'WEEKEND?': (352, 0, 576, 32),
    'WHAT': (576, 0, 704, 32),
    'DID': (704, 0, 800, 32),
    'YOU': (800, 0, 896, 32),
    'DO?': (896, 0, 992, 32)
}

# Bounding box coordinates for each word in the lowercase sentence
bounding_boxes_lowercase = {
    'hi,': (-288, 0, -192, 32),
    'lisa!': (-192, 0, -32, 32),
    'did': (-32, 0, 64, 32),
    'you': (64, 0, 160, 32),
    'finish': (160, 0, 352, 32),
    'the': (352, 0, 448, 32),
    'book': (448, 0, 576, 32),
    'i': (576, 0, 608, 32),
    'lent': (608, 0, 736, 32),
    'you?': (736, 0, 864, 32)
}

# Calculate the plot limits for uppercase and lowercase sentences
xmin_upper = min(bbox[0] for bbox in bounding_boxes_uppercase.values())
xmax_upper = max(bbox[2] for bbox in bounding_boxes_uppercase.values())
ymin_upper = min(bbox[1] for bbox in bounding_boxes_uppercase.values())
ymax_upper = max(bbox[3] for bbox in bounding_boxes_uppercase.values())

xmin_lower = min(bbox[0] for bbox in bounding_boxes_lowercase.values())
xmax_lower = max(bbox[2] for bbox in bounding_boxes_lowercase.values())
ymin_lower = min(bbox[1] for bbox in bounding_boxes_lowercase.values())
ymax_lower = max(bbox[3] for bbox in bounding_boxes_lowercase.values())

# Plotting
plt.figure(figsize=(18, 6))

# Plotting for uppercase sentence
plt.subplot(1, 2, 1)
plt.title('Landing Spot (uppercase)')
for word, bbox in bounding_boxes_uppercase.items():
    xmin, ymin, xmax, ymax = bbox
    plt.gca().add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, linewidth=2, edgecolor='r', facecolor='none'))
    plt.text((xmin + xmax) / 2, (ymin + ymax) / 2, word, horizontalalignment='center', verticalalignment='center', fontsize=12)

plt.scatter(fixation_coordinates_uppercase['FPOGX'] * (xmax_upper - xmin_upper) + xmin_upper, fixation_coordinates_uppercase['FPOGY'] * (ymax_upper - ymin_upper) + ymin_upper, color='blue', marker='o', s=100, label='Fixation Point')
plt.xlim(xmin_upper - 50, xmax_upper + 50)
plt.ylim(ymin_upper - 5, ymax_upper + 5)
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)

# Plotting for lowercase sentence
plt.subplot(1, 2, 2)
plt.title('Landing Spot (lowercase)')
for word, bbox in bounding_boxes_lowercase.items():
    xmin, ymin, xmax, ymax = bbox
    plt.gca().add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, linewidth=2, edgecolor='r', facecolor='none'))
    plt.text((xmin + xmax) / 2, (ymin + ymax) / 2, word, horizontalalignment='center', verticalalignment='center', fontsize=12)

plt.scatter(fixation_coordinates_lowercase['FPOGX'] * (xmax_lower - xmin_lower) + xmin_lower, fixation_coordinates_lowercase['FPOGY'] * (ymax_lower - ymin_lower) + ymin_lower, color='blue', marker='o', s=100, label='Fixation Point')
plt.xlim(xmin_lower - 50, xmax_lower + 50)
plt.ylim(ymin_lower - 5, ymax_lower + 5)
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)

# Add a super title with the subject's name
plt.suptitle('Subject 11', fontsize=16, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust the layout to make room for the suptitle
plt.show()
```
