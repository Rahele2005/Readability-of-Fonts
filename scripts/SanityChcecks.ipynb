```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv("/Users/svevabattisti/Desktop/MA in Linguistics/classes/I SEMESTER/Eye Tracking/FINAL REPORT /VS CODE/data/subject-11.csv")

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# List of logs to check
logs_to_check = [
    'SENTENCE_LOWERCASE_ONSET',
    'SENTENCE_LOWERCASE_OFFSET_BOOKS',
    'SENTENCE_UPPERCASE_ONSET',
    'SENTENCE_UPPERCASE_OFFSET_WEEKEND',
    'SENTENCE_LOWERCASE_OFFSET_WEEKEND',
    'SENTENCE_UPPERCASE_OFFSET_BOOKS'
]

# Ensure the 'USER' column exists
if 'USER' in df.columns:
    # Get unique values in the 'USER' column
    user_column_values = df['USER'].unique()
    
    # Check which logs are present in the 'USER' column
    presence = [1 if log in user_column_values else 0 for log in logs_to_check]
    
    # Plot the presence of logs
    plt.figure(figsize=(10, 6))
    plt.bar(logs_to_check, presence, color='blue', alpha=0.7)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Logs')
    plt.ylabel('Presence (1 = Present, 0 = Missing)')
    plt.title('Presence of Specified Logs in USER Column')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Print missing logs for reference
    missing_logs = [log for log, present in zip(logs_to_check, presence) if not present]
    if not missing_logs:
        print("All specified logs are present in the 'USER' column.")
    else:
        print("The following logs are missing from the 'USER' column:")
        for log in missing_logs:
            print(f"- {log}")
else:
    print("The 'USER' column does not exist in the DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to load and clean the CSV data
def load_and_clean_data(file_path):
    df = pd.read_csv(file_path)
    df.columns = df.columns.str.strip()  # Strip whitespace from column names
    return df

# Path to the CSV file
file_path = "/Users/svevabattisti/Desktop/MA in Linguistics/classes/I SEMESTER/Eye Tracking/FINAL REPORT /VS CODE/data/subject-11.csv"

# Load the data
df = load_and_clean_data(file_path)

# Function to plot FPOGX against USER log/triggers
def plot_fpogx_vs_triggers(subject_id, data):
    plt.figure(figsize=(8, 4))  # Adjusting the figure size to make it smaller
    
    # Plot FPOGX against sample index (or time)
    plt.plot(data.index, data['FPOGX'], label=f'{subject_id} FPOGX', color='blue')
    
    # Add vertical lines and text annotations for USER log/triggers
    for i, row in data.iterrows():
        if row['USER'] in ['SENTENCE_LOWERCASE_ONSET', 'SENTENCE_LOWERCASE_OFFSET_BOOKS', 'SENTENCE_UPPERCASE_ONSET', 'SENTENCE_UPPERCASE_OFFSET_WEEKEND']:
            plt.axvline(x=i, color='red', linestyle='--')
            plt.text(i, data['FPOGX'].max(), row['USER'], rotation=90, verticalalignment='bottom', color='red')
    
    plt.xlabel('Sample Index')
    plt.ylabel('FPOGX')
    plt.title(f'FPOGX vs USER log/triggers for {subject_id}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot FPOGX vs USER log/triggers for the loaded subject
subject_id = "subject-11"
plot_fpogx_vs_triggers(subject_id, df)


```
